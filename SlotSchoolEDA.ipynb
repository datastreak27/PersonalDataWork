#### Primary Research Question: Are gambling venues deliberately concentrated near schools?


```python
# For this reasearch question, I will utilize the gamblingkeywords dataframe, gamblingplace_keywords dataframe and the school_place_proximity dataframe

# The gamblingkeywords dataframe contains the keywords associated with gambling places (eg. "bets, "slots, "casino"), I will not use this dataframe since 
# we are instredted in the gambling places themselves and not the keywords associated with them, which can be resolved via an SQL script later on.
# The school_place_proximity dataframe contains the proximity of schools to gambling places, which includes the distance in meters between the school and the gambling place.

# 

# Extract all raw Distance values
distances = school_place_proximity_df["Distance"]
# Define bins by every 5000 meters
max_d = int(distances.max())
bins = np.arange(0, max_d + 5000, 5000)
# Compute counts per bin
counts, edges = np.histogram(distances, bins=bins)
# Plot bar chart of all distances
plt.figure(figsize=(10, 6))
plt.bar(edges[:-1], counts, width=np.diff(edges), edgecolor="black", align="edge")
plt.xlabel("Distance to Nearest School (m)")
plt.ylabel("Number of School–Gambling Venue Pairs")
plt.title("School–Gambling Venue Distances")
plt.xticks(edges, rotation=45)
plt.tight_layout()
plt.show()


```
